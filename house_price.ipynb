```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
# Simple data
x = np.array([[1000, 2, 5],
              [1500, 3, 2],
              [1200, 2, 10],
              [1800, 4, 3],
              [800, 1, 15]], dtype=float)

y = np.array([50, 70, 45, 85, 30], dtype=float)  # prices in lakhs
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(3,)),  # 3 features
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1)  # final output: price
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x,y, epochs=100)
```

```python
# Predict for a new house: [1300 sqft, 3 bedrooms, 5 years old]
new_data = scaler.transform([[1300, 3, 5]])
predicted_price = model.predict(new_data)
print(f"Predicted Price: ₹{predicted_price[0][0]:.2f} lakhs")
```
